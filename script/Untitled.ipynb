## Linear Polynomial

### Basis Function

```python
for n1_phi in [0, 2]:

    x_phi = [0, 1, 2, 3, 4]
    y_phi = [0, 0, 0, 0, 0]
    y_phi[n1_phi] = 1

    labels = [f'x{i}' for i in range(5)]

    with plt.xkcd():
        fig, ax = plt.subplots(
            figsize=(9, 3),
            tight_layout=True
        )

        ax.plot(
            x_phi, y_phi
        )
        ax.plot(
            [x_phi[n1_phi]]*2, [0, 1],
            linestyle="--"
        )
        ax.scatter(
            x_phi[n1_phi], 1,
            color="black",
            zorder=3
        )

        ax.annotate(
            f'THE  BASIC  FUNCTION\nPHI_{n1_phi}  DEFINED  IN  X_{n1_phi}',
            xy=(x_phi[n1_phi]+0.1, 1), 
            arrowprops=dict(arrowstyle='->'), 
            xytext=(x_phi[n1_phi]+0.75, 0.75)
        )

        fig.text(
            0.5, -0.1,
            '"The Mountain Function" from One dimensional linear element',
            ha='center',
            fontsize=18
        )

        ax.spines["right"].set_color('none')
        ax.spines["top"].set_color('none')

        ax.set_xticks(x_phi)
        ax.set_xticklabels(labels)
        ax.set_yticks([0, 1])

        ax.set_xlabel("Space", fontsize=15)
        ax.set_ylabel("Value", fontsize=15)

#     plt.savefig(f"basis_fun_{n1_phi}.png")
```

```python
labels = [f'x{i}' for i in range(5)]

with plt.xkcd():
    fig, axs = plt.subplots(
        5, 1,
        figsize=(9, 12),
        sharex="col",
        tight_layout=True
    )

    for n1_phi in range(5):
        x_phi = [0, 1, 2, 3, 4]
        y_phi = [0, 0, 0, 0, 0]
        
        y_phi[n1_phi] = 1
        
        ax = axs[n1_phi]
        ax.plot(
            x_phi, y_phi
        )
        ax.plot(
            [x_phi[n1_phi]]*2, [0, 1],
            linestyle="--"
        )
        ax.scatter(
            x_phi[n1_phi], 1,
            color="black",
            zorder=3
        )

        ax.annotate(
            f'THE  BASIC  FUNCTION\nPHI_{n1_phi}  DEFINED  IN  X_{n1_phi}',
            xy=(x_phi[n1_phi]+0.1, 1) if n1_phi < 2 else (x_phi[n1_phi]-0.1, 1), 
            arrowprops=dict(arrowstyle='->'), 
            xytext=(x_phi[n1_phi]+0.75, 0.75) if n1_phi < 2 else (x_phi[n1_phi]-2.1, 0.75),
            color="gray"
        )

        ax.spines["right"].set_color('none')
        ax.spines["top"].set_color('none')

        ax.set_xticks(x_phi)
        ax.set_xticklabels(labels)
        ax.set_yticks([0, 1])

        ax.set_xlabel("Space", fontsize=15)
        ax.set_ylabel("Value", fontsize=15)
        
    ax.text(
        2, -0.75,
        '"The Mountain Function" from One dimensional linear element',
        ha='center',
        fontsize=18,
    )

# plt.savefig(f"basis_fun_all.png")    
```

### Iteration

```python
labels = [f'x{i}' for i in range(5)]

with plt.xkcd():
    fig, ax = plt.subplots(
        figsize=(9, 6),
        sharex="col",
        tight_layout=True
    )
    
    ax.set_ylim(-0.02, 1.01)

    x_phi = [0, 1, 2, 3, 4, 5]
    y_phi = [0.1, 0.5, 0.8, 0.6, 0.2, 0.4]
    
    f = lagrange(x_phi, y_phi)

    ax.plot(
        x_phi, y_phi,
        color='tab:blue',
        # marker='.',
        # markersize=12,
        # markeredgecolor='tab:red',
        # markerfacecolor='tab:red',
        zorder=3
    )
    
    for i, j in zip(x_phi, y_phi):
        ax.plot(
            [i, i], [0, j],
            linestyle='--',
            color='tab:gray',
        )
    
    ax.scatter(
        x_phi[:-1], y_phi[:-1],
        s=42,
        color='tab:orange',
        zorder=4
    )
    
    ax.scatter(
        x_phi[-1], y_phi[-1],
        s=60,
        color='tab:red',
        zorder=4
    )
    
    ax.plot(
        np.linspace(0, 5, 101),
        f(np.linspace(0, 5, 101)),
        color="black"
    )
    
    ax.text(
        x_phi[-1]-1.55, y_phi[-1]+0.22,
        'UNKNOWN POINTS',
        fontsize=18,
        color="gray"
    )

    ax.annotate(
        'f(x5, y(x5))',
        xy=(x_phi[-1], y_phi[-1]+0.01), 
        arrowprops=dict(arrowstyle='->'), 
        xytext=(x_phi[-1]-1., y_phi[-1]+0.15),
        color="black"
    )

    ax.spines["right"].set_color('none')
    ax.spines["top"].set_color('none')

    ax.set_xticks(x_phi)
    ax.set_xticklabels(labels+['x5'])
    ax.set_yticks([0, 1])

    ax.set_xlabel("x", fontsize=15)
    ax.set_ylabel("f(x, y(x))", fontsize=15)

    ax.text(
        2.5, -0.25,
        'How to integrate with Unknown Point?',
        ha='center',
        fontsize=18,
    )

plt.savefig(f"unknown_point_yn1.png") 
```

## Quadratic Polynomial

```python
def get_basis_quad(x_seq):
    return (
        lambda x: (x - x_seq[1]) * (x - x_seq[2]) / (x_seq[0] - x_seq[1]) / (x_seq[0] - x_seq[2]),
        lambda x: (x - x_seq[0]) * (x - x_seq[2]) / (x_seq[1] - x_seq[0]) / (x_seq[1] - x_seq[2]),
        lambda x: (x - x_seq[0]) * (x - x_seq[1]) / (x_seq[2] - x_seq[0]) / (x_seq[2] - x_seq[1])
    )

# ------------------------------------
with plt.xkcd():
    fig, ax = plt.subplots(
        figsize=(9, 6),
        sharex="col",
        tight_layout=True
    )
    
    x_range = np.linspace(0, 2, 100)
    f1, f2, f3 = get_basis_quad([0, 1, 2])
    
    ax.plot(
        x_range, f1(x_range),
        linestyle='-', alpha=0.75, color='tab:blue'
    )
    ax.plot(
        x_range, f2(x_range),
        linestyle='-', alpha=0.75, color='tab:orange'
    )
    ax.plot(
        x_range, f3(x_range),
        linestyle='-', alpha=0.75, color='tab:green'
    )
    
    ax.scatter(
        [0, 1, 2], [1, 1, 1],
        s=45, color='tab:red',
        zorder=3
    )
    
    ax.plot(
        [0, 0], [0, 1],
        alpha=0.3,
        linestyle='--',
        color='tab:blue'
    )
    
    ax.plot(
        [1, 1], [0, 1],
        alpha=0.3,
        linestyle='--',
        color='tab:orange'
    )
    
    ax.plot(
        [2, 2], [0, 1],
        alpha=0.3,
        linestyle='--',
        color='tab:green'
    )
    
    ax.spines["right"].set_color('none')
    ax.spines["top"].set_color('none')

    ax.set_xticks([0, 1, 2])
    ax.set_xticklabels(['x0', 'x1', 'x2'])
    ax.set_yticks([0, 1])

    ax.set_xlabel("x", fontsize=15)
    ax.set_ylabel("f(x, y(x))", fontsize=15)

    ax.text(
        1.0, -0.45,
        'How to integrate with Unknown Point?',
        ha='center',
        fontsize=18,
    )

plt.savefig(f"2-order_basis.png") 
```
